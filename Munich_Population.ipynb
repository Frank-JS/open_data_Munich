# Bevölkerungsentwicklung in München seit 1900

Datenquelle: https://www.opengov-muenchen.de/dataset/bevoelkerung

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pop = pd.read_csv("pop.csv")
print("Zeilen / Spalten im Datensatz:", pop.shape)
pop.head()
```

```python
# Jahr als Indexspalte
pop.index = pop["jahr"]
# Felder im Jahr 1900 für einwohner_deutsche" und 'einwohner_auslaender_innen' mit Punkt (".") --> ersetzen durch None
pop.at[1900, "einwohner_deutsche"] = 0
pop.at[1900, 'einwohner_auslaender_innen'] = 0
# Spalten für einwohner_deutsche" und 'einwohner_auslaender_innen' von str --> int umwandeln
pop["einwohner_deutsche"] = pop["einwohner_deutsche"].astype(int)
pop['einwohner_auslaender_innen'] = pop['einwohner_auslaender_innen'].astype(int) 
```

```python
# neue Spalten berechnen für Anzeile Ausländer, Männer
pop["Anteil_Ausländer"] = pop['einwohner_auslaender_innen'] / pop['einwohner_Insgesamt']
pop["Anteil_Männer"] = pop['einwohner_maennlich'] / pop['einwohner_Insgesamt']
```

```python
pop[['einwohner_Insgesamt', 'einwohner_maennlich', 'einwohner_weiblich', 'einwohner_auslaender_innen']].plot()
```

```python
pop[["Anteil_Ausländer", "Anteil_Männer"]].plot()
```
